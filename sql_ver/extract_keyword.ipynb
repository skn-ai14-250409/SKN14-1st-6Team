```python
data['리콜사유'] = data['리콜사유'].str.replace("현대자동차","")
data['리콜사유'] = data['리콜사유'].str.replace("㈜","")
# data['리콜사유'] = data['리콜사유'].str.replace("포터2(HR)","")
# data['리콜사유'] = data['리콜사유'].str.replace("있습니다.","")
# data['리콜사유'] = data['리콜사유'].str.replace("의도치 않게","")
# data['리콜사유'] = data['리콜사유'].str.replace("제작결함시정","")
# data['리콜사유'] = data['리콜사유'].str.replace("하고자 함","")

data
```

```python
keyword = []
for i in data['리콜사유'].tolist()[:20]:
    kw_model = KeyBERT()
    # keywords = kw_model.extract_keywords(i,keyphrase_ngram_range=(2,6),use_maxsum = True,top_n = 1)
    keywords_mmr = kw_model.extract_keywords(i,keyphrase_ngram_range=(2,6),use_mmr = True,top_n = 1,diversity = 0.5)
    keyword.append(keywords_mmr[0][0])
```
